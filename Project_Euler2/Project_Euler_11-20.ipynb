### Problem 12
#### Find the first Triangle number that has over 500 factors

```python
#Produce triangle number, check how many factors is has
def factor_check(x):
    my_hand = []
    beer = 1
    while beer not in my_hand:
        if x%beer == 0:
            my_hand.append(beer)
            my_hand.append(x/beer)
        beer += 1
    return(len(my_hand))


triangle = 3
add = 3
while factor_check(triangle) <= 200:
    triangle += add
    add += 1
print(triangle)

```

### Problem 14
#### Find longest chain of Collatz sequence that starts under 1 million

```python
#if n is even n = n/2
#n = 3n+1 if n is odd
long_number = 0
longest = 0
for i in range(1,10**6):
    current = 0
    number = i
    while i != 1:
        if i%2 == 0:
            i = i/2
            current += 1
        else:
            i = 3*i + 1
            current += 1
    
    if current > longest:
        longest = current
        long_number = number
print(longest)
print(long_number)




```

### Problem 18
#### Longest Path

```python

triangle = []
with open("Project_Euler_Problem18.txt", "r") as file:
    for line in file:
        row = list(map(int, line.split()))
        triangle.append(row)
print(triangle)

for i in range(len(triangle) - 2, -1, -1):
    for j in range(len(triangle[i])):
        triangle[i][j] += max(triangle[i + 1][j], triangle[i + 1][j + 1])

max_sum = triangle[0][0]
print("The maximum path sum is:", max_sum)
```

### Problem 19
#### How many Sundays fell on the first of the month during the twentieth century

```python
thirty_one_days = [1,3,5,7,8,10,0]
thirty_days = [4,6,9,11]
year = 1901
day = 2 #was a tuesday in 1901
month = 1
count = 0
#Check day 1 of january, if day%7 = 0, then add 1
#Then for january add 31 days, check if day%7 = 0...

while year < 2001:
    if day%7 == 0:
        count += 1
    if month%12 in thirty_one_days:
        day += 31
        month += 1
    elif month%12 in thirty_days:
        day += 30
        month +=1 
    else:
        if year%4 == 0:
            day += 29
            month += 1
        else:
            day += 28
            month += 1
    if month%12 == 0:
        year += 1

print(count)
        

```
